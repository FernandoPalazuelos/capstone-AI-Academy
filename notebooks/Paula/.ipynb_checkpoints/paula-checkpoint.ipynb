Data Cleaning process

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing, svm
from sklearn.preprocessing import MaxAbsScaler
from sklearn.linear_model import LogisticRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split, cross_val_score, learning_curve
from sklearn.metrics import accuracy_score
from sklearn.metrics import explained_variance_score, mean_squared_error, confusion_matrix
```

```python
# Loading the data
url = 'https://raw.githubusercontent.com/GerardoA01701474/capstone-AI-Academy/master/data/breastCancer.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.columns
```

```python
data.dtypes
```

```python
#Data cleaning in bare_nucleoli '?'
data['bare_nucleoli'].value_counts()
```

```python
#Convert and change str to int for those values who isn't '?' [symbol]
data = data[data['bare_nucleoli']!='?']
data['bare_nucleoli'] = data['bare_nucleoli'].astype(int)
```

```python
data['bare_nucleoli'].value_counts()
```

```python
# Drop 'id' column 
data.drop(['id'], axis=1, inplace=True)
data.head()
```

```python
#Creating correlation matrix
matrix = data.corr().round(1)

# Correlation heat map for correlation matrix
sns.heatmap(matrix, annot=True)
plt.title("Heat map Correlation Matrix \n")
plt.show()
```

```python
#Drop unnecesary columns - low correlation 
data.drop(['mitoses'], axis=1, inplace=True)
data.head()
```

```python
#Drop class (since it's string)
boxplot_data = data.drop(['class'], axis=1)
# Box plot - outliners 
plt.figure(figsize=(15,5))
boxplot_data.plot.box(grid=False, rot=45, fontsize=8)
plt.tight_layout()
plt.title("Box plot")
```

```python
# Outliners checkout 
boxplot_data.describe().round(2)
```

```python
# Change int to a str column 
data['class'] = data['class'].astype(str)
```

```python
# class 2 for benign, 4 for malignant
data['class'].value_counts()
```

```python
# ploting class ---
x_values = data['class'].unique()
y_values = data['class'].value_counts().tolist()

# x axis new names - plot 
x_names = ['Benign', 'Malignant']

# plot
ax = plt.subplot()
plt.bar(x_values,y_values, color = ['green', 'purple'])
plt.title('Benign vs. Malignant')
ax.set_xlabel("Class")
plt.xticks(x_values, x_names)
plt.show()
```

```python
#Density plot
data.plot.kde()
plt.title("Density")
```

```python
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10,20))  # Adjust the layout based on your preference

axes = axes.ravel()  # Flattens the axis array

for i, column in enumerate(data.columns):
    data[column].plot.kde(ax=axes[i])
    axes[i].set_title(column)

plt.tight_layout()
plt.show()
```

## Model 

```python
df_x = data[['clump_thickness', 'size_uniformity', 'shape_uniformity',
       'marginal_adhesion', 'epithelial_size', 'bare_nucleoli',
       'bland_chromatin', 'normal_nucleoli']] 

df_y = data["class"]
```

```python
########################### splitting data into train and test ###############################
Xtrain, Xtest, ytrain, ytest = train_test_split(df_x.values, df_y.values,random_state=1)
model = LogisticRegression(fit_intercept = True)
model.fit(Xtrain,ytrain)
pred_y = model.predict(Xtest)
acc = accuracy_score(ytest, pred_y)
```

```python
print("accuracy of the model: "+ str(acc))
print("score of the cross validation: " + str(cross_val_score(model,df_x,df_y, cv=10).mean()))
print("Matrix of confusion: \n" + str(confusion_matrix(ytest, pred_y)))
```

## Test-Train Results

```python
#import the libary to perform classification report
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Predict on training data
train_predictions = model.predict(Xtrain)

# Predict on testing data
test_predictions = model.predict(Xtest)

train_accuracy = accuracy_score(ytrain, train_predictions)
test_accuracy = accuracy_score(ytest, test_predictions)

# Print results
print("Training Results:")
print("Accuracy:", train_accuracy)
print(classification_report(ytrain, train_predictions))

print("Testing Results:")
print("Accuracy:", test_accuracy)
print(classification_report(ytest, test_predictions))
```

## Some highlights from the results 


support:  the number of data that each metric was calculated on.

f1 score: F1 score represents the Harmonic Mean of Precision and Recall. F1 score is generally the most used metric for describing the performance of a model.

accuracy:  It provides a solid holistic view of the overall performance of our model. Is useful because it allows us to measure the total number of predictions a model gets right, including both True Positives and True Negatives

Weighted average: considers the performance of each class but gives more weight to classes with more samples. This means that the performance of larger classes has a bigger impact on the overall average.

Macro Average: the average precision, recall and f1 score between classes. Macro average doesn't take class imbalance into the calculation. The macro average calculates the average performance across all classes by considering each class equally.
So if you do have class imbalance, pay attention to this metric.
(the avg performace from each metric)

## Overfitting evaluation

```python
train_accuracy = model.score(Xtrain, ytrain)
test_accuracy = model.score(Xtest, ytest)

#Calculating the difference between the train and test accuracy 
overfitting_score = train_accuracy - test_accuracy

#Printing result 
print("Overfitting Score:", round(overfitting_score,2))
```

```python
scores = cross_val_score(model, df_x, df_y, cv=10)
print(f"Cross-validated scores: {scores}")
```

```python
#Learning Curves: Useful for visualizing how adding more training data affects performance.
from sklearn.model_selection import learning_curve

train_sizes, train_scores, val_scores = learning_curve(model, df_x, df_y, train_sizes=np.linspace(0.1, 1.0, 10))
train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)

plt.plot(train_sizes, train_scores_mean, label='Training score')
plt.plot(train_sizes, val_scores_mean, label='Cross-validation score')
plt.title('Learning Curve')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.legend()
```
